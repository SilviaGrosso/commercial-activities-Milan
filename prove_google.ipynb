```python
#!pip install googlemaps pandas

import googlemaps
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import plotly.express as px
```

```python
from config import API_KEY

# Utilizza la tua chiave API
api_key = API_KEY
lat_numb = 45.4654219
long_numb = 9.1859243
```

```python
categories = [
    'grocery', 
    'supermarket',
    'bakery',
    'butcher',
    'convenience_store',
    'bar',
    'cafe',
    "alcohol",
    "beverages",
    "brewing_supplies",
    "butcher",
    "cheese",
    "chocolate",
    "coffee",
    "confectionery",
    "convenience",
    "deli",
    "dairy",
    "farm",
    "frozen_food",
    "greengrocer",
    "health_food",
    "ice_cream",
    "pasta",
    "pastry",
    "seafood",
    "spices",
    "tea",
    "wine",
    "water",
    "food",
    "farm" ]
```

```python
#center of milan
center_lat = lat_numb
center_lng = long_numb
```

```python
df_list = []

for cat in categories:
    gmaps = googlemaps.Client(key = api_key)

    places_result = gmaps.places_nearby(
        location=(center_lat, center_lng),
        radius=50000,
        keyword= cat
        )
    for place in places_result['results']:
        name = place['name']
        lat = place['geometry']['location']['lat']
        lng = place['geometry']['location']['lng']

        df_list.append({'Nome': name, 'Latitudine': lat, 'Longitudine': lng, 'Categoria': cat})

#df with all points
google_markets = pd.DataFrame(df_list)
```
