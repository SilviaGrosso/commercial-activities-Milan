```python
categories = [
    'grocery', 
    'supermarket',
    'bakery',
    'butcher',
    'convenience_store',
    'bar',
    'cafe',
    "alcohol",
    "beverages",
    "brewing_supplies",
    "butcher",
    "cheese",
    "chocolate",
    "coffee",
    "confectionery",
    "convenience",
    "deli",
    "dairy",
    "farm",
    "frozen_food",
    "greengrocer",
    "health_food",
    "ice_cream",
    "pasta",
    "pastry",
    "seafood",
    "spices",
    "tea",
    "wine",
    "water",
    "food",
    "farm" ]
```

```python
# #center of milan
# center_lat = lat_numb
# center_lng = long_numb
```

```python
df_list = []

for cat in categories:
    gmaps = googlemaps.Client(key = api_key)

    for i in range(0,len(lat)):
        center_lat = lat[i]
        center_lng = long[i]

        places_result = gmaps.places_nearby(
            location=(center_lat, center_lng),
            radius=10000,
            keyword= cat
            )
        for place in places_result['results']:
            name = place['name']
            latit = place['geometry']['location']['lat']
            longit = place['geometry']['location']['lng']

            df_list.append({'Nome': name, 'Latitudine': latit, 'Longitudine': longit, 'Categoria': cat})

#df with all points
google_markets = pd.DataFrame(df_list)
```

```python
len(google_markets)
```

```python
google_markets = google_markets.drop_duplicates(keep='first')
```

```python
df = google_markets

color_scale = [(0, 'orange'), (1,'red')]

fig = px.scatter_mapbox(df, 
                        lat="Latitudine", 
                        lon="Longitudine", 
                        hover_name="Nome",
                        color_continuous_scale=color_scale,
                        zoom=8, 
                        height=800,
                        width=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
