```python
lat = [45.522526, 45.530935, 45.496016, 45.474849, 45.475615, 45.455599, 45.454323, 45.445648, 45.429313, 45.410165, 45.439012, 45.420074, 45.406088, 
45.420308, 45.442474, 45.450422, 45.473058, 45.480039,45.490148, 45.489667, 45.498090,45.515895, 45.513490, 45.516376, 45.516136]

long = [9.231373, 9.194635, 9.244103, 9.236101, 9.264109, 9.234717, 9.264543, 9.207800,  9.224532, 9.238718, 9.185248,  9.182506, 9.183909, 
 9.151980, 9.151980, 9.140651, 9.127948,  9.096362, 9.142367, 9.109752,  9.138934, 9.126918, 9.144427, 9.172236, 9.201762] 
```

```python
df_list = []

for cat in categories:
    gmaps = googlemaps.Client(key = api_key)

    for i in range(0,len(lat)):
        center_lat = lat[i]
        center_lng = long[i]

        places_result = gmaps.places_nearby(
            location=(center_lat, center_lng),
            radius=10000,
            keyword= cat
            )
        for place in places_result['results']:
            name = place['name']
            latit = place['geometry']['location']['lat']
            longit = place['geometry']['location']['lng']

            df_list.append({'Nome': name, 'Latitudine': latit, 'Longitudine': longit, 'Categoria': cat})

#df with all points
google_markets = pd.DataFrame(df_list)
```

```python
google_markets = google_markets.drop_duplicates(keep='first')
```
