# Milan urban food policy pact since Expo 2015 

Equity, resilience and sustainability articulated in its social, economic, and environmental components.
Through the Food Policy, the City of Milan is working to orientate the choices that directly or indirectly concern food and water within the framework of its institutional competences and in the activities of the associated companies, in order to improve people's quality of life and the quality of the territory and to play an innovative role at a national and international level.

At the heart of food policy is education in schools, to strengthen not only the link with the land but also with those who work and produce the food.






The five Food Policy priorities are:

-	ensure access to healthy food for all
-	promoting the sustainability of the food system
-	educating people on food
-	fighting waste
-	supporting and promoting local agri-food research


From 2015 to 2021, thanks to the change in school menus, where vegetable proteins were favoured, Milan calculated a 42.89% cut in CO2 equivalent emissions.


## Food Logistics

Food logistics is one of the components of the food system that is difficult to analyse on an urban scale, partly due to a lack of data and public research. It is, however, a very relevant topic: a 2014 AMAT survey, focusing on food logistics in the Lombardy capital, showed that food accounts for 43% of goods handling activities. Today, the incidence of urban goods traffic has increased incredibly, given the constant growth of e-commerce and, specifically, food delivery, which imply deliveries with smaller loads, more frequent, capillary and bound by precise schedules determined by the availability and needs of the final recipients.

### 5 Food waste hub - Earthshot Prize in 2021 

Model of recovery and redistribution of food surpluses based on local neighbourhood networks.

The actions started in 2018/2019 with the launch of a first pilot project in Municipio 9 and are now replicated in other neighbourhoods of the city.



### Hub logistic model
Hubs are intermediares between **supermarkets** and **onlus**. They make it possible to recover food surpluses from supermarkets, store them and redistribute them to the most needy people through the collaboration of local associations. 

This logic enable the possibility to act on every food type, including fresh food, and providing onlus with a high diversity of meals that wouldn't be possible to obtain otherwise.

## What is missing?
While the main focus of the Milan food policy is on a **centralized strategy** to reduce food waste, there is a lack of a **decentralized approach** to the problem (i.e. the food waste hub model doesn't involve the citizens directly).

Intoduce a strategy to involve citizens in the process of reducing the food waste can bring several benefits:
- **increase the awareness** of the citizens on this problem (that is fundamental considering that the 42% of the food waste is generated by the households)
- **bring economic benefits** both for the citizens and for the producers (i.e. the citizens can buy the food at a lower price and the producers can sell the food that would be wasted)
- **involve local shop** in the process of reducing the food waste (i.e. the citizens can buy the food that would be wasted from the local shop). This seems to be very important considering that local shops are a fundamental part of the city identity and they are struggling to survive.

### How to do it?

A possible direction is to **involve private company** specialized in the food waste fight (i.e. **Too Good To Go**) to create a platform that can connect the citizens with the local shops and the producers. 

For example, the idea behind Too Good To Go is brilliant but it lacks in terms of usability. The platform is not intuitive and there is no way to search for a specific type of product. 

A **possible solution** is to create an application that can compute the best route to visit the local shops and buy the food that would be wasted. As well as, the application can show the benefits in terms of money saved, food saved and CO2 saved. If the route is chosen to be travelled by bike or by foot, the application can also show the positive impact on the health of the citizens.

This kind of strategy can achieve the goals previously described, and lead toward:
- **a more sustainable city**
- **a more aware city**
- **a more healthy city**
- **a more inclusive city**
